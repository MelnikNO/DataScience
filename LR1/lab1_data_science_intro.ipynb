## Задание 3: Формулирование гипотез и анализ

### 3.1 Базовая статистика

```python
# Посчитайте средний возраст пассажиров
average_age = df['Age'].mean()
print(f"Средний возраст пассажиров: {average_age:.2f} лет")

# Посчитайте медианный возраст
median_age = df['Age'].median()
print(f"Медианный возраст пассажиров: {median_age:.2f} лет")
```

```python
# Посчитайте количество мужчин и женщин
gender_counts = df['Sex'].value_counts()
print("Распределение по полу:")
print(gender_counts)

# В процентах
gender_percentages = df['Sex'].value_counts(normalize=True) * 100
print("\nРаспределение по полу (в процентах):")
print(gender_percentages.round(2))
```

```python
# Посчитайте общий процент выживших
survival_rate = df['Survived'].mean() * 100
print(f"Общий процент выживших: {survival_rate:.2f}%")

# Процент выживших по полу
survival_by_gender = df.groupby('Sex')['Survived'].mean() * 100
print("\nПроцент выживших по полу:")
print(survival_by_gender.round(2))
```

### 3.2 Формулирование гипотез

На основе полученных данных сформулируйте и проверьте простые гипотезы:

**Гипотеза 1:** Женщины имели больше шансов на выживание, чем мужчины.

**Ваша проверка:**

```python
# Проверьте гипотезу 1
women_survival = df[df['Sex'] == 'female']['Survived'].mean() * 100
men_survival = df[df['Sex'] == 'male']['Survived'].mean() * 100

print(f"Процент выживших женщин: {women_survival:.2f}%")
print(f"Процент выживших мужчин: {men_survival:.2f}%")
print(f"Разница: {women_survival - men_survival:.2f} процентных пунктов")

if women_survival > men_survival:
    print("\nГипотеза ПОДТВЕРЖДАЕТСЯ: женщины действительно имели больше шансов на выживание.")
else:
    print("\nГипотеза НЕ ПОДТВЕРЖДАЕТСЯ.")
```

**Гипотеза 2:** Пассажиры первого класса имели больше шансов на выживание.

**Ваша проверка:**

```python
# Проверьте гипотезу 2
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print("Процент выживших по классам:")
for pclass, survival in survival_by_class.items():
    print(f"Класс {pclass}: {survival:.2f}%")

print(f"\nРазница между 1-м и 3-м классом: {survival_by_class[1] - survival_by_class[3]:.2f} процентных пунктов")
```

### Задание 6: Создайте свою визуализацию

```python
# Попробуйте создать свой график
# Например, выживаемость по классам
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100

plt.figure(figsize=(8, 6))

# Создаем столбчатую диаграмму
bars = plt.bar(survival_by_class.index, survival_by_class.values, 
               color=['#FFD93D', '#6BCB77', '#4D96FF'], 
               alpha=0.8, edgecolor='black', linewidth=1.5)

# Настройка заголовков и подписей
plt.title('Процент выживших по классам', fontsize=16, fontweight='bold')
plt.xlabel('Класс', fontsize=13)
plt.ylabel('Процент выживших (%)', fontsize=13)
plt.ylim(0, 100)
plt.xticks([1, 2, 3], ['1-й класс', '2-й класс', '3-й класс'])

# Добавляем значения на столбцы
for bar, value in zip(bars, survival_by_class.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1.5,
             f'{value:.1f}%', ha='center', va='bottom', 
             fontsize=12, fontweight='bold')
    
    # Добавляем количество пассажиров под каждым столбцом
    count = len(df[df['Pclass'] == bar.get_x()])
    plt.text(bar.get_x() + bar.get_width()/2, -5,
             f'n={count}', ha='center', va='bottom', 
             fontsize=10, color='gray')

# Добавляем сетку
plt.grid(True, alpha=0.3, axis='y')

# Показываем график
plt.tight_layout()
plt.show()

# Вывод статистики в консоль
print("\n📊 СТАТИСТИКА ПО КЛАССАМ:")
print("-" * 50)
for pclass, survival in survival_by_class.items():
    count = len(df[df['Pclass'] == pclass])
    survived = len(df[(df['Pclass'] == pclass) & (df['Survived'] == 1)])
    print(f"Класс {pclass}: {survival:.1f}% выживших ({survived} из {count} пассажиров)")
```

## Анализ EDA и методологии CRISP-DM на примере данных Титаника

В ходе самостоятельного изучения практик разведочного анализа данных (EDA) на платформе Kaggle и анализа Jupyter-ноутбуков других специалистов были выявлены ключевые закономерности и подходы, применяемые при работе с датасетом Титаника. Результаты этого исследования позволяют четко определить место EDA в рамках методологии CRISP-DM.

### 🔍 Место EDA в методологии CRISP-DM

Методология CRISP-DM (Cross-Industry Standard Process for Data Mining) — это стандартизированный подход к организации работы с данными, состоящий из шести ключевых этапов: понимание бизнеса, понимание данных, подготовка данных, моделирование, оценка и внедрение .

Анализ показывает, что разведочный анализ данных (EDA) не является изолированным этапом, а применяется на нескольких этапах CRISP-DM, при этом наиболее активно — на этапах "Понимание данных" и "Подготовка данных".

**1. Этап "Понимание данных" (Data Understanding)** — это основной этап для применения EDA .

На данном этапе EDA используется для:
- **Первичного исследования структуры данных**: изучение количества записей, типов данных, наличия пропусков .
- **Визуализации распределений**: построение гистограмм для числовых признаков (возраст, стоимость билета) и столбчатых диаграмм для категориальных (пол, класс) .
- **Выявления взаимосвязей**: анализ выживаемости в разрезе различных признаков (пол, класс, порт посадки) .
- **Визуализации пропусков**: использование библиотек для наглядного представления пропущенных значений .

**2. Этап "Подготовка данных" (Data Preparation)** — здесь EDA помогает обосновать решения по очистке и трансформации данных .

На этом этапе EDA позволяет:
- **Принимать решения об обработке пропусков**: например, заполнение возраста медианой на основе анализа распределения .
- **Создавать новые признаки**: на основе анализа существующих данных (например, выделение фамилий из имени, создание признака размера семьи) .
- **Обосновывать удаление признаков**: например, признак Cabin часто удаляется из-за большого количества пропусков .

**3. Этап "Оценка" (Evaluation)** — EDA также применяется для интерпретации результатов моделирования, например, анализ важности признаков и визуализация ошибок модели .

### 📊 Анализ типов данных и выбор диаграмм

Изучение практик EDA показывает, что выбор типа визуализации напрямую зависит от типа данных в колонке:

**Числовые (непрерывные) признаки**:
- *Примеры*: Age, Fare
- *Рекомендуемые диаграммы*: гистограммы, boxplot, violin plot для анализа распределения .
- *Цель*: выявление выбросов, оценка асимметрии, сравнение распределений.

**Категориальные (бинарные) признаки**:
- *Примеры*: Sex, Survived
- *Рекомендуемые диаграммы*: столбчатые диаграммы, круговые диаграммы .
- *Цель*: анализ соотношения категорий, сравнение выживаемости.

**Категориальные (ординальные) признаки**:
- *Примеры*: Pclass (1, 2, 3)
- *Рекомендуемые диаграммы*: столбчатые диаграммы, boxplot с группировкой .
- *Цель*: выявление закономерностей в упорядоченных категориях.

**Категориальные (номинативные) признаки**:
- *Примеры*: Embarked, Cabin
- *Рекомендуемые диаграммы*: столбчатые диаграммы.
- *Цель*: анализ распределения по категориям, выявление редких значений.

### 💡 Инструменты для EDA

Анализ Jupyter-ноутбуков показывает, что для EDA активно используются следующие библиотеки:

- **Pandas**: для базовой работы с таблицами, группировки, агрегации данных.
- **Matplotlib**: для построения базовых графиков.
- **Seaborn**: как надстройка над Matplotlib для создания более эстетичных и информативных визуализаций .
- **Plotly**: для создания интерактивных визуализаций.
- **Missingno**: для визуализации пропущенных значений .

### 📝 Выводы

Разведочный анализ данных (EDA) является критически важным этапом в методологии CRISP-DM, особенно на этапах "Понимание данных" и "Подготовка данных". Он позволяет обоснованно принимать решения о методах очистки, трансформации и выборе признаков для моделирования. Правильный выбор типов диаграмм в зависимости от типа данных позволяет эффективно выявлять закономерности, проверять гипотезы и готовить данные для построения качественных моделей машинного обучения.
